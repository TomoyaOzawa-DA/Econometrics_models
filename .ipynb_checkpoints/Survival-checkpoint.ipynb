# Survival analysis

## Estimating the survival function using Kaplan-Meier curve

- Tutorial from [lifelines](https://lifelines.readthedocs.io/en/latest/Survival%20analysis%20with%20lifelines.html#estimating-the-survival-function-using-kaplan-meier) library



```python
import pandas as pd
from lifelines.datasets import load_dd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = load_dd()
data.head()
```

Set up:
- birth: start of the individual's tenure
- death: end of their tenure
- Censor: end of this data set (2008) or die while thier tenure

```python
from lifelines import KaplanMeierFitter
kmf = KaplanMeierFitter()
```

```python
T = data["duration"]
E = data["observed"] # takes 1 if their death is observed

kmf.fit(T, event_observed=E)
```

```python
# plot for survival function
kmf.survival_function_.plot()
plt.title('Survival function of political regimes')
```

```python
# with confidence interval
kmf.plot_survival_function()
```

```python
data["democracy"].unique()
```

```python
ax = plt.subplot(111)

dem = (data["democracy"] == "Democracy")

kmf.fit(T[dem], event_observed=E[dem], label="Democratic Regimes")
kmf.plot_survival_function(ax=ax)

kmf.fit(T[~dem], event_observed=E[~dem], label="Non-democratic Regimes")
kmf.plot_survival_function(ax=ax)

plt.title("Lifespans of different global regimes")
```

```python
# we can adjust the timeline as follows
ax = plt.subplot(111)

t = np.linspace(0, 50, 51)
kmf.fit(T[dem], event_observed=E[dem], timeline=t, label="Democratic Regimes")
ax = kmf.plot_survival_function(ax=ax)

kmf.fit(T[~dem], event_observed=E[~dem], timeline=t, label="Non-democratic Regimes")
ax = kmf.plot_survival_function(ax=ax)

plt.title("Lifespans of different global regimes");
```

```python
regime_types = data['regime'].unique()
enumerate(regime_types)
```

```python
# when we have multiple groups:
regime_types = data['regime'].unique()

for i, regime_type in enumerate(regime_types):
    ax = plt.subplot(2, 3, i + 1)

    ix = data['regime'] == regime_type
    kmf.fit(T[ix], E[ix], label=regime_type)
    kmf.plot_survival_function(ax=ax, legend=False)

    plt.title(regime_type)
    plt.xlim(0, 50)

    if i==0:
        plt.ylabel('Frac. in power after $n$ years')

plt.tight_layout()
```

## Cox-ph model

```python
from lifelines import CoxPHFitter
from lifelines.datasets import load_rossi
```

```python
df = load_rossi()

cph = CoxPHFitter()
cph.fit(df, 'week', 'arrest')
```

```python
df.head()
```

```python
cph.print_summary()
cph.plot();
```

```python
# predict
result = cph.predict_survival_function(df.iloc[0])
```

```python
result
```

```python
def compute_VaR(dat, model_hedonic, model_cph, prob):
    # input
    # dat: データフレーム, 予測したい物件のスペック, 現状は1物件のみに対応
    # model_hedonic: 学習済みのヘドニック賃金関数
    # model_cph: 学習済みのcox model
    # prob: 信頼水準
    
    # output:
    # VaR: 信頼水準 prob %におけるValue at Risk
    
    # 予測したい物件について生存関数を推定
    S_t = model_cph.predict_survival_function(dat) # インデックスに空室期間, 各行に対応する生存関数の値が格納されているデータフレーム
    
    # S(t) = 1-probに近いtを抽出する
    target = 1 - prob
    taregt_total_days = abs(S_t - target).idxmin()[0]# 目標の値からの差分が一番小さいS_tのインデックス（=total_days）を取得
    
    #VaRを計算する
    D = dat.total_money/ 30 # 1日あたりの逸失家賃
    VaR = D * taregt_total_days 
    
    return VaR
```

```python
fig, ax = plt.subplots()

ax.plot(result.index, result.iloc[:, 0])
```

### Asccessing the cox-ph model by residuals

**Martingale residuals**

```python
r = cph.compute_residuals(df, 'martingale')
r.head()
```

```python
r.plot.scatter(
    x='week', y='martingale', c=np.where(r['arrest'], '#008fd5', '#fc4f30'),
    alpha=0.75
)
```

- observed sample
    - most of people have longer duration to get arrested than expected according to this model since their matringale residuals are postive in many cases

**Deviance residuals**

```python
r = cph.compute_residuals(df, 'deviance')
r.head()
```

```python
r.plot.scatter(
    x='week', y='deviance', c=np.where(r['arrest'], '#008fd5', '#fc4f30'),
    alpha=0.75
)
```
